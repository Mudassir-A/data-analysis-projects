```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the CSV dataset into pandas dataframe
data = pd.read_csv('./titanic.csv')
```

```python
data
```

```python
# cleaning of the dataset
data = data[data['Age'].notna()]
```

```python
# Create a histogram of the age column
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic')
plt.show()
```

```python
# Generate the line plot to visualize the trend of the 'Fare' variable over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='PassengerId', y='Fare', data=data)
plt.xlabel('Passenger IDs')
plt.ylabel('Fare')
plt.title('Fare Trend over Time')
plt.show()
```

```python
# create a bar chart to show the frequency of categories in the 'Pclas' column
plt.figure(figsize=(8, 8))
sns.countplot(x='Pclass', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Passenger Class Distribution on the Titanic')
plt.show()
```

```python
# Design a scatterplot to explore the relationship between 'Age' and 'Fare'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=data, hue='Sex')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Relationship between Age and Fare')
plt.show()
```

```python
# Use a pie-chart to represent the composition of the 'Sex' xolumn
sex_counts = data['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Composition on the Titanic')
plt.show()
```

```python
# Build a box plot to visualize the distribution of the 'Fare' variable
plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=data)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```
